```python
# Import Modules
import pandas as pd
import matplotlib.pyplot as plt
import plotly 
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
import numpy as np 
from datetime import datetime as dt 
```

```python
# Load data
data = pd.read_excel("/Users/talhajamal/Desktop/Code/Regression_Analysis/data/data_coursework_Q2.xls")
data["Yt_USR"] = (data['USR'] > 0).astype(int)
data["Yt_ESR"] = (data['ESR'] > 0).astype(int)
data["Yt_1_USR"] = data["Yt_USR"].shift(1)
data["Yt_1_ESR"] = data["Yt_ESR"].shift(1)
data.dropna(inplace=True)
data.head(10)
```

```python
# USR OLS
USR_X = data[['UIF', 'URR', 'UDY', 'UOIL']]
USR_X = sm.add_constant(USR_X)
USR_Y = data['Yt_1_USR']
USR_Model = sm.OLS(USR_Y, USR_X).fit()
print(USR_Model.summary())
```

```python
# ESR OLS
ESR_X = data[['EIF', 'ERECB', 'EFX', 'EDY', 'ESP']]
ESR_X = sm.add_constant(ESR_X)
ESR_Y = data['Yt_1_ESR']
ESR_Model = sm.OLS(ESR_Y, ESR_X).fit()
print(ESR_Model.summary())
```
